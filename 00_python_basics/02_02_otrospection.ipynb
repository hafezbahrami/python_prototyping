```python
def class_stub(_class, include_func=True):
    stub = f"class {_class.__name__}:\n"
    stub += f'    """{_class.__doc__}"""\n\n'
    # stub += "\n".join([f"    {k}: {v}" for k, v in _class.__annotations__.items()])
    stub += "\n" 

    if include_func:
        functions = [v for _, v in _class.__dict__.items() if callable(v)]
        for func in functions:
            # _args = ", ".join([f"{k}: {v}" for k, v in func.__annotations__.items()])
            # stub += f"    def {func.__name__}({_args}):\n"
            stub += f'        """{func.__doc__}"""'

    return stub

print("-" * 120)
print(class_stub(np.float32))
print("-" * 120)
```
